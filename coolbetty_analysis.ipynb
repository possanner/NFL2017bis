```python
PPG_ar = []
YPG_ar = []

i = 0
for team in team_list:  
    PPG = (1 - fac)*data_dict[team]['PPG_OFF'] + fac*data_dict[team]['PPG_OFF_0']
    PPG_ar.append(PPG) 
    
    YPG = (1 - fac)*data_dict[team]['YPG_OFF'] + fac*data_dict[team]['YPG_OFF_0']
    YPG_ar.append(YPG) 
    
PPG_mean = np.mean(PPG_ar)
YPG_mean = np.mean(YPG_ar)
PPY = PPG_mean/YPG_mean
print('points per yard:', PPY)

for team in team_list:
    data_dict[team]['PPG_OFF_2'] = (1 - fac)*data_dict[team]['YPG_OFF']*PPY + fac*data_dict[team]['YPG_OFF_0']*PPY
    data_dict[team]['PPG_OFF_home_2'] = (1 - fac)*data_dict[team]['YPG_OFF_home']*PPY + fac*data_dict[team]['YPG_OFF_0_home']*PPY
    data_dict[team]['PPG_OFF_away_2'] = (1 - fac)*data_dict[team]['YPG_OFF_away']*PPY + fac*data_dict[team]['YPG_OFF_0_away']*PPY
    
    data_dict[team]['PPG_DEF_2'] = (1 - fac)*data_dict[team]['YPG_DEF']*PPY + fac*data_dict[team]['YPG_DEF_0']*PPY
    data_dict[team]['PPG_DEF_home_2'] = (1 - fac)*data_dict[team]['YPG_DEF_home']*PPY + fac*data_dict[team]['YPG_DEF_0_home']*PPY
    data_dict[team]['PPG_DEF_away_2'] = (1 - fac)*data_dict[team]['YPG_DEF_away']*PPY + fac*data_dict[team]['YPG_DEF_0_away']*PPY 
```

```python
OFF_ar = []
DEF_ar = []
pow_ar = []

OFF_home_ar = []
DEF_home_ar = []
pow_home_ar = []

OFF_away_ar = []
DEF_away_ar = []
pow_away_ar = []

coeff_ar = np.ndarray((32, 2))

i = 0    
for team in team_list:
    
    # total:
    rnk_1 = (1 - fac)*data_dict[team]['PPG_OFF'] + fac*data_dict[team]['PPG_OFF_0']
    rnk_2 = data_dict[team]['PPG_OFF_2']
    OFF_val = 2/3*rnk_1 + 1/3*rnk_2
    rnk_3 = (1 - fac)*data_dict[team]['PPG_DEF'] + fac*data_dict[team]['PPG_DEF_0']
    rnk_4 = data_dict[team]['PPG_DEF_2']
    DEF_val = 2/3*rnk_3 + 1/3*rnk_4
     
    OFF_ar.append([OFF_val, team, rnk_1, rnk_2])
    DEF_ar.append([DEF_val, team, rnk_3, rnk_4])
    pow_ar.append([OFF_val - DEF_val, team, rnk_1 - rnk_3, rnk_2 - rnk_4])
    
    # home:
    rnk_5 = (1 - fac)*data_dict[team]['PPG_OFF_home'] + fac*data_dict[team]['PPG_OFF_0_home']
    rnk_6 = data_dict[team]['PPG_OFF_home_2']
    OFF_home_val = 2/3*rnk_5 + 1/3*rnk_6
    rnk_7 = (1 - fac)*data_dict[team]['PPG_DEF_home'] + fac*data_dict[team]['PPG_DEF_0_home']
    rnk_8 = data_dict[team]['PPG_DEF_home_2']
    DEF_home_val = 2/3*rnk_7 + 1/3*rnk_8
    
    OFF_home_ar.append([OFF_home_val, team, rnk_5, rnk_6])
    DEF_home_ar.append([DEF_home_val, team, rnk_7, rnk_8])
    pow_home_ar.append([OFF_home_val - DEF_home_val, team, rnk_5 - rnk_7, rnk_6 - rnk_8])
    
    # away:
    rnk_9 = (1 - fac)*data_dict[team]['PPG_OFF_away'] + fac*data_dict[team]['PPG_OFF_0_away']
    rnk_10 = data_dict[team]['PPG_OFF_away_2']
    OFF_away_val = 2/3*rnk_9 + 1/3*rnk_10
    rnk_11 = (1 - fac)*data_dict[team]['PPG_DEF_away'] + fac*data_dict[team]['PPG_DEF_0_away']
    rnk_12 = data_dict[team]['PPG_DEF_away_2']
    DEF_away_val = 2/3*rnk_11 + 1/3*rnk_12
    
    OFF_away_ar.append([OFF_away_val, team, rnk_9, rnk_10])
    DEF_away_ar.append([DEF_away_val, team, rnk_11, rnk_12])
    pow_away_ar.append([OFF_away_val - DEF_away_val, team, rnk_9 - rnk_11, rnk_10 - rnk_12])
    
    # home field advantage:
    coeff_ar[i][0] = OFF_home_val - DEF_home_val
    coeff_ar[i][1] = OFF_away_val - DEF_away_val
    i += 1

# total:
power_OFF = sorted(OFF_ar, reverse=True)
power_DEF = sorted(DEF_ar, reverse=False)
power_rankings = sorted(pow_ar, reverse=True)

print('Power rankings:           rating    scoring      yds    ')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_rankings[i][1].rjust(22), '{0:+5.1f}'.format(power_rankings[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_rankings[i][2], power_rankings[i][3]))
    team = power_rankings[i][1].lstrip().rstrip()
    data_dict[team]['power'] = [i+1, power_rankings[i][0]]
    
print()
print('Offense:                  rating    scoring      yds    ')
print()

for i in range(32):
    print('{:2d}'.format(i + 1), power_OFF[i][1].rjust(22), ' {0:4.1f}'.format(power_OFF[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_OFF[i][2], power_OFF[i][3]))
    team = power_OFF[i][1].lstrip().rstrip()
    data_dict[team]['OFFENSE'] = [i+1, power_OFF[i][0]]
    
print()
print('Defense:                  rating    scoring      yds    ')
print()

for i in range(32):
    print('{:2d}'.format(i + 1), power_DEF[i][1].rjust(22), ' {0:4.1f}'.format(power_DEF[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_DEF[i][2], power_DEF[i][3]))
    team = power_DEF[i][1].lstrip().rstrip()
    data_dict[team]['DEFENSE'] = [i+1, power_DEF[i][0]]
    
# home:
power_OFF_home = sorted(OFF_home_ar, reverse=True)
power_DEF_home = sorted(DEF_home_ar, reverse=False)
power_rankings_home = sorted(pow_home_ar, reverse=True)

print()
print('HOME Power rankings:     rating    scoring      yds    ')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_rankings_home[i][1].rjust(22), '{0:+5.1f}'.format(power_rankings_home[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_rankings_home[i][2], power_rankings_home[i][3]))
    
print()
print('HOME Offense:            rating    scoring      yds    ')
print()

for i in range(32):
    print('{:2d}'.format(i + 1), power_OFF_home[i][1].rjust(22), ' {0:4.1f}'.format(power_OFF_home[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_OFF_home[i][2], power_OFF_home[i][3]))
    team = power_OFF_home[i][1].lstrip().rstrip()
    data_dict[team]['OFFENSE_home'] = [i+1, power_OFF_home[i][0]]
    
print()
print('HOME Defense:            rating    scoring      yds    ')
print()

for i in range(32):
    print('{:2d}'.format(i + 1), power_DEF_home[i][1].rjust(22), ' {0:4.1f}'.format(power_DEF_home[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_DEF_home[i][2], power_DEF_home[i][3]))
    team = power_DEF_home[i][1].lstrip().rstrip()
    data_dict[team]['DEFENSE_home'] = [i+1, power_DEF_home[i][0]]
    
# away:
power_OFF_away = sorted(OFF_away_ar, reverse=True)
power_DEF_away = sorted(DEF_away_ar, reverse=False)
power_rankings_away = sorted(pow_away_ar, reverse=True)

print()
print('AWAY Power rankings:     rating    scoring      yds    ')
print()
for i in range(32):
    print('{:2d}'.format(i + 1), power_rankings_away[i][1].rjust(22), '{0:+5.1f}'.format(power_rankings_away[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_rankings_away[i][2], power_rankings_away[i][3]))
    
print()
print('AWAY Offense:            rating    scoring      yds    ')
print()

for i in range(32):
    print('{:2d}'.format(i + 1), power_OFF_away[i][1].rjust(22), ' {0:4.1f}'.format(power_OFF_away[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_OFF_away[i][2], power_OFF_away[i][3]))
    team = power_OFF_away[i][1].lstrip().rstrip()
    data_dict[team]['OFFENSE_away'] = [i+1, power_OFF_away[i][0]]
    
print()
print('AWAY Defense:            rating    scoring      yds    ')
print()

for i in range(32):
    print('{:2d}'.format(i + 1), power_DEF_away[i][1].rjust(22), ' {0:4.1f}'.format(power_DEF_away[i][0]), '     {0:4.1f}       {1:4.1f}   '.format(power_DEF_away[i][2], power_DEF_away[i][3]))
    team = power_DEF_away[i][1].lstrip().rstrip()
    data_dict[team]['DEFENSE_away'] = [i+1, power_DEF_away[i][0]]
```
